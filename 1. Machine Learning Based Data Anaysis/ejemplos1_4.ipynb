## Tratamiento de datos categóricos

```python
df = pd.DataFrame([['green', 'M', 10.1, 'class2'],
                   ['red', 'L', 13.5, 'class1'],
                   ['blue', 'XL', 15.3, 'class2']])
df
```

```python
df.columns = ['color', 'talla', 'peso', 'clase'] # Asignar nombres a las columnas
talla_mapping = {'M':1, 'L':2, 'XL':3} # Diccionario de mapeo, asignando números a las tallas
df['talla'] = df['talla'].map(talla_mapping) # Aplicar el mapeo a la columna 'talla'
df
```

```python
clase_mapping = {'class1':0, 'class2':1} # Diccionario de mapeo para la clase
df['clase'] = df['clase'].map(clase_mapping) # Aplicar el mapeo a la columna 'clase'
colores_mapping = {'red':0, 'green':1, 'blue':2} # Diccionario de mapeo para colores
df['color'] = df['color'].map(colores_mapping) # Aplicar el mapeo a la columna 'color'
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # Crear el codificador de etiquetas basado en sklearn con un ordenamiento automático
df = pd.DataFrame([['green', 'M', 10.1, 'class2'],
                   ['red', 'L', 13.5, 'class1'],
                   ['blue', 'XL', 15.3, 'class2']])
df.columns = ['color', 'talla', 'peso', 'clase']

df['color'] = le.fit_transform(df['color'])
df['talla'] = le.fit_transform(df['talla'])
df['clase'] = le.fit_transform(df['clase'])
df
```

## One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data = [['red', 'M', 10.1, 'class2'],
        ['red', 'L', 13.5, 'class1'],   
        ['blue', 'XL', 15.3, 'class2']]
df = pd.DataFrame(data, columns=['color', 'size', 'value', 'class'])
print("DataFrame original:")
print(df)
df_dummies = pd.get_dummies(df, columns=['color', 'size', 'class'])
print("DataFrame con variables categóricas convertidas en variables dummy:")
print(df_dummies)
# OneHotEncoder de sklearn
ohe = OneHotEncoder(sparse_output=False)
#ohe_fit = ohe.fit_transform(df[['color', 'size', 'class']])
encoded_df = pd.DataFrame(ohe.fit_transform(df[['color', 'size', 'class']]), columns=ohe.get_feature_names_out(['color', 'size', 'class']))
encoded_df = pd.concat([ df[['value']].reset_index(drop=True), encoded_df], axis=1)
print("DataFrame con variables categóricas convertidas en variables dummy (OneHotEncoder):")
print(encoded_df)
```
